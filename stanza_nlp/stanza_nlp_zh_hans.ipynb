```
!pip install stanza

import stanza
stanza.download('zh-hans')
```

```
zh_nlp = stanza.Pipeline('zh')
```

```
# data = "" #@param {type:"string"}
```

```
zh_file = open('zh_text.txt', 'r')
data = zh_file.read()
print(data)
```

```
zh_doc = zh_nlp(data)
for i, sent in enumerate(zh_doc.sentences):
    print("[Sentence {}]".format(i+1))
    print("{:16s}\t{:16s}\t{:6s}\t{:6s}\t{:16s}".format(\
        'Text', 'Lemma', 'Pos', 'Head', 'Deprel'))
    for word in sent.words:
        print("{:16s}\t{:16s}\t{:6s}\t{:d}\t{:16s}".format(\
            word.text, word.lemma, word.pos, word.head, word.deprel))
    print("")
```

```
for i, sent in enumerate(zh_doc.sentences):
    spaced = ''
    for word in sent.words:
        spaced = spaced + word.text + ' '
    print(spaced, '\n')
```
